```python
# Import important library
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_name = "M182081.csv"
# Read raw file
df = pd.read_csv(file_name)
print("Preview first 20 rows:")
print(df.head(20))
```

New Section

```python
# Drop the first 9 rows
df = df.drop(index=range(0, 8)).reset_index(drop=True)
print("\nPreview after dropping first 10 rows:")
print(df.head(10))
```

```python
# Preview the last 105 rows to see the file structure
print("\nPreview last 105 rows:")
print(df.tail(105))
```

```python
# Keep only the first 171 rows (based on inspection)
df = df.iloc[:171].reset_index(drop=True)
df
```

```python
# Use the first row as header
df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)

print("\nData after setting header row:")
print(df.head(10))
```

```python
df.shape
```

```python
# Preview column name
print(df.columns)
```

```python

# Standardize column names (remove spaces and special characters)
df.columns = (
    df.columns.astype(str)
    .str.strip()
    .str.replace(' ', '_')
    .str.replace('[^0-9a-zA-Z_]', '', regex=True)
)

print("\n Cleaned column names:")
print(df.columns.tolist())
```

```python
# Identify year columns
year_cols = [col for col in df.columns if col.isdigit()]
```

```python
# Drop columns before 2000
old_years = [col for col in year_cols if int(col) < 2000]
df = df.drop(columns=old_years)
print(f"\n Dropped {len(old_years)} columns before 2000: {old_years}")
```

```python
# Keep only 2000–2024 (reconfirm)
year_cols = [col for col in df.columns if col.isdigit() and 2000 <= int(col) <= 2024]
non_year_cols = [col for col in df.columns if col not in year_cols]
df = df[non_year_cols + year_cols]

print(f"\n Final check: kept {len(year_cols)} columns (2000–2024).")
print(f"Remaining years: {year_cols}")
```

```python
print([col for col in df.columns if col.isdigit() and int(col) < 2000])
```

```python
print(df.dtypes)
```

```python
# Check and handle missing values
print("\n Missing values before filling:")
print(df.isnull().sum())
```

```python
# na   not available or not applicable
# nec  not elsewhere classified
# nes  not elsewhere specified
#  -     nil or negligible or not significant

# Replace special notations
# '-' → 0
# 'na', 'n.a.', 'n.a', 'NA' → np.nan → fill 0
for col in year_cols:
    df[col] = (
        df[col]
        .replace({'-': 0, 'na':0})
        .astype(float)
    )

print("\n Replaced special symbols ('-', 'na', etc.) and filled missing with 0.")
```

```python
print(df.isnull().sum())
```

```python
# Drop completely empty rows (if any)
df.dropna(how='all', inplace=True)
```

```python
# Remove duplicates if exist
df.drop_duplicates(inplace=True)
```

```python

# Reset index
df = df.reset_index(drop=True)
print(f"\n Data cleaned successfully. Final shape: {df.shape}")
```

```python
occ_set = {
    re.sub(r'[^a-z0-9]+', ' ', str(x).lower()).strip()
    for x in [
        "All Occupation Groups, (Total Employed Residents)",  
        "Managers & Administrators (Including Working Proprietors)",
        "Professionals",
        "Technicians & Associate Professionals",
        "Associate Professionals & Technicians",              
        "Clerical Support Workers",
        "Service & Sales Workers",
        "Craftsmen & Related Trades Workers",
        "Plant & Machine Operators & Assemblers",
        "Cleaners, Labourers & Related Workers"              
    ]
}
```

```python
# Scan to assign occupation + industry
text_col = "Data_Series"
current_occ = None
occ_list, ind_list = [], []


for raw in df[text_col].astype(str):
    s = raw.strip()
    if s == "":
        occ_list.append(np.nan)
        ind_list.append(np.nan)
        continue

    cleaned = re.sub(r'[^a-z0-9]+', ' ', str(s).lower()).strip()
    
    if cleaned in occ_set:
        current_occ = s
        occ_list.append(current_occ)
        ind_list.append("All Industries")
    else:
        if current_occ is None:
            current_occ = "All Occupation Groups (Total Employed Residents)"
        occ_list.append(current_occ)
        ind_list.append(s)

df["occupation"] = occ_list
df["industry"] = ind_list
```

```python
# Drop blanks
df = df.dropna(subset=["occupation"]).reset_index(drop=True)
```

```python
#  Melt to tidy format
df = df.melt(
    id_vars=["occupation", "industry"],
    value_vars=year_cols,
    var_name="year",
    value_name="employment"
)

df["year"] = df["year"].astype(int)
df["employment"] = pd.to_numeric(df["employment"], errors="coerce").fillna(0.0)
```

```python
df["occupation"] = df["occupation"].str.replace(r"\s+", " ", regex=True).str.strip()
df["industry"] = df["industry"].str.replace(r"\s+", " ", regex=True).str.strip()
```

```python
# Preview
print("\n Data successfully reshaped into tidy format:")
print(df.head(12))
print(f"\nShape: {df.shape}")
print("Unique occupations:", df['occupation'].nunique())
print("Unique industries:", df['industry'].nunique())
```

```python
# Preview cleaned data
print("\nPreview of cleaned data (first 10 rows):")
print(df.head(10))
```

```python
# Save cleaned file
output_path = "unemployment_by occupation & industry.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")
print(f"\n File saved: {output_path}")
```
